```python
meg_pico = pd.DataFrame.from_dict(meg_pico, orient='index', columns=['count'])
```

```python
meg_pico.to_csv('meg_pico.csv')
```

```python
meg_pico
```

```python
meg_pico['Date'] = meg_pico.index
meg_pico.set_index('Date', inplace=True)
meg_pico.head()
```

```python
def plot_series(meg_pico, cols=None, title='Title', xlab=None, ylab='Count (unit = readings)', steps=1):
    
    # Set figure size to be (18, 9).
    plt.figure(figsize=(18,9))
    
    # Iterate through each column name.
    for col in cols:
            
        # Generate a line plot of the column name.
        # You only have to specify Y, since our
        # index will be a datetime index.
        plt.plot(meg_pico[col])
        
    # Generate title and labels.
    plt.title(title, fontsize=26)
    plt.xlabel(xlab, fontsize=20)
    plt.ylabel(ylab, fontsize=20)
    
    # Enlarge tick marks.
    plt.yticks(fontsize=18)
    plt.xticks(meg_pico.index[0:steps], rotation=75, fontsize=18)
    grid(color='gainsboro', linestyle=':', linewidth=2);
```

```python
# Generate a time plot of our data.
plot_series(meg_pico, ['count'], title = 'Counts of Purim-Night Megillah Readings at Selected Times', steps = 1000)
```

```python
# Plot ACF
plot_acf(meg_pico['count'], lags = 20);
```

### There does not appear to be statistically significant trend for the night time Megillah times, but it there does appear to be some seaonality.

```python
# Plot PACF
plot_pacf(meg_pico['count'], lags = 20);
```

### There appears to be seaonality throughout, most notibly with the last 5 lags
